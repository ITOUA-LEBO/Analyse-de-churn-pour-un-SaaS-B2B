# 📉 Analyse de churn – SaaS B2B
Ce notebook explore les comportements utilisateurs à risque de churn.


```python
# 📦 Import des librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
sns.set(style='whitegrid')
```

```python
# 📂 Chargement des données (exemple fictif)
# df = pd.read_csv('data/churn_dataset.csv')
df = pd.DataFrame({
    'user_id': range(1, 6),
    'login_freq': [3, 1, 4, 0, 2],
    'support_tickets': [0, 3, 1, 5, 2],
    'feature_usage': [0.8, 0.3, 0.9, 0.2, 0.5],
    'churned': [0, 1, 0, 1, 0]
})
df.head()
```

```python
# 🔍 Visualisation de la corrélation avec le churn
corr = df.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Corrélation entre variables')
plt.show()
```
